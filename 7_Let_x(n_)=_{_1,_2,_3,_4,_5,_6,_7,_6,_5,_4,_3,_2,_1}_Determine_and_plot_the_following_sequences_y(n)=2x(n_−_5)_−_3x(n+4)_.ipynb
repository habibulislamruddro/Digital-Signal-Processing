```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x={
    -5:1,
    -4:2,
    -3:3,
    -2:4,
    -1:5,
    0:6,
    1:7,
    2:6,
    3:5,
    4:4,
    5:3,
    6:2,
    7:1,
}
```

```python
y_index=np.arange(-10,15,1)
```

```python
y=np.array([])
```

```python
for i in y_index:
    temp_1 = 0
    temp_2 = 0
    if x.get(i-5) is not None:
        temp_1 = x[i-5]
    if x.get(i+4) is not None:
        temp_2 = x[i+4]

    temp = 2*temp_1 - 3*temp_2
    y = np.append(y,temp)
```

```python
plt.plot(y_index,y)
plt.grid()
plt.show()
```
