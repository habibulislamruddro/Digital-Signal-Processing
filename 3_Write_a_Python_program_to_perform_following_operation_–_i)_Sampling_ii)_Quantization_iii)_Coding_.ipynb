```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
analog_f=100
analog_a=5
sampling_rate=10*analog_f
```

```python
def analog_signal(f,a,x):
    y=a*np.sin(2*np.pi*f*x)
    return y
```

```python
x=np.arange(0,1/analog_f,1/sampling_rate)
y=analog_signal(analog_f,analog_a,x)
```

```python
plt.figure(figsize=(20,10))
plt.subplot(2,2,1)
plt.plot(x,y)
plt.title("Analog Signal")
plt.grid()
```

```python
x=np.arange(0,1/analog_f,1/sampling_rate)
y=analog_signal(analog_f,analog_a,x)
```

```python
plt.subplot(2,2,2)
plt.stem(x,y)
plt.title("Sampled Signal")
plt.grid()
```

```python
def quantize(x):
    y=np.zeros(len(x))
    #for i in range(len(x)):
      #  y[i]=int(x[i])
    y=np.round(x).astype(int)
    return y
```

```python
y=quantize(y)
```

```python
plt.subplot(2,2,3)
plt.stem(x,y)
plt.title("Quantized Signal")
plt.grid()
```

```python
def coding(x):
    code=""
    for i in range(len(x)):
        temp=bin(x[i])
        if(x[i]<0):
            temp=temp[3:]
            while(len(temp)<3):
                temp='0'+temp
            temp='1'+temp
        else:
            temp=temp[2:]
            while(len(temp)<3):
                temp='0'+temp
            temp='0'+temp
        code=code+temp
    return code
```

```python
code=coding(y)
```

```python
x=np.arange(len(code))
y=np.zeros(len(code))
for i in range(len(code)):
    if(code[i]=='1'):
        y[i]=1
    else:
        y[i]=-1
```

```python
plt.subplot(2,2,4)
plt.step(x,y,where="post")
plt.title("Digital Signal")
plt.grid()

plt.show()
```
